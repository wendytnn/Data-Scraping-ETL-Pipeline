# Data Scrapping 

#### **** The data scrapping process need around 50s to 1 minutes  running time due to large amount datasets ****

- The data science job listing  is collected from a career website called careerjet. 
- The specific countries that I selected for scrapping is United States where it have all the cities location resides in the United States. 
- The core informations is collected from the CareerJet Websites which are 
1. Job Title 
2. Company name
3. Location (The cities in United States)
4. Date which the job was posted. 
5. Salary (Salary Range)


```python
jobwebsite = mechanicalsoup.Browser()

jobs_list = []
company_list = []
locations_list = []
descriptions_list = []
date_joblisted_list = []
salary_list = []

for page_number in range(1, 100):
    url = f'https://www.careerjet.com/data-science-jobs.html?p={page_number}'
    
    
    page = jobwebsite.get(url)
    soup = BeautifulSoup(page.content, 'lxml')


    careerjet = page.soup

    job_title = careerjet.select('h2 a')
    companies = careerjet.select('.company')
    locations = careerjet.select('.location li')
    date_joblisted = soup.select('.badge-r')
    salary = soup.select('ul.salary li')

        
    for job_list in range(len(job_title)):
        jobs_list.append(job_title[job_list].text.strip() if job_list < len(job_title) else "Non Avaialable")
        company_list.append(companies[job_list].text.strip() if job_list < len(companies) else "Non Avaialable")
        locations_list.append(locations[job_list].text.strip() if job_list < len(locations) else "Non Avaialble")
        date_joblisted_list.append(date_joblisted[job_list].text.strip() if job_list < len(date_joblisted) else "Non Avaialable")
        salary_list.append(salary[job_list].text.strip() if job_list < len(salary) else 0)
        



print(jobs_list)
print(company_list)
print(locations_list)
print(date_joblisted_list)
print(salary_list)


```

# Data cleaning and Feature Engineering 
## Used OOP Classes

- Clean the Salary column to get the Average Salary for each jobs 
- Perform feature engineering to create a new list and name it as Average Salary


#### The Average Salary column is created to run the followinf query in the SQL stated in the CA2 projects 

### --- List average pay/salary ---


```python
class Average_Salary:
    def __init__(self):
        pass

    def get_average_salaries(self, salary_list):
        average_salaries = []

        for salary in salary_list:
            salary_str = str(salary)
            salary = [float(salary.replace(',', '')) for salary in re.findall(r'\d{1,3}(?:,\d{3})*(?:\.\d+)?', salary_str)]

            if salary:
                average_salary = sum(salary) / len(salary)
                average_salaries.append(average_salary)

        return average_salaries

average = Average_Salary()
average_salaries = average.get_average_salaries(salary_list)
print(average_salaries)
```

##  CREATE CLASSES 

### Perform matching or grouping to identify 
- Maximum Salary
- Minimum Salary
- Hourly Salary
- Interval or Period of Salary 

```python


class Salary:
    def __init__(self):
        self.min_salary_list = []
        self.max_salary_list = []
        self.hour_salary_list = []
        self.period_list = []

    def get_salary(self, salary):
        find = re.search(r'\$([\d,]+(?:\.\d{2})?)-?([\d,]+(?:\.\d{2})?)?\sper\s(year|hour)?', str(salary))
        if find:
            minimum_salary = find.group(1).replace(',', '') if find.group(1) else None
            maximum_salary = find.group(2).replace(',', '') if find.group(2) else None
            period = find.group(3) if find.group(3) else None

            if period == 'hour' and not minimum_salary:
                minimum_salary = None

            return minimum_salary, maximum_salary, period
        else:
            return None, None, None

    def salary_info(self, salary_list):
        self.min_salary_list.clear()
        self.max_salary_list.clear()
        self.hour_salary_list.clear()
        self.period_list.clear()

        salary_info = [self.get_salary(s) for s in salary_list]

        for info in salary_info:
            salary_min = info[0]
            self.min_salary_list.append(salary_min)
            salary_max = info[1]
            self.max_salary_list.append(salary_max)
            period = info[2]
            self.period_list.append(period)

            salary_min = float(salary_min) if salary_min is not None else None
            salary_max = float(salary_max) if salary_max is not None else None

            if period == "hour":
                self.hour_salary_list.append(int(salary_min) if salary_min is not None else None)
            else:
                self.hour_salary_list.append(None)

    def dataframe_salary(self):
        salary_df = pd.DataFrame({
            'Minimum Salary': self.min_salary_list,
            'Maximum Salary': self.max_salary_list,
            'Interval': self.period_list
        })
        return salary_df



salary = Salary()
salary.salary_info(salary_list)

print(salary.min_salary_list)
print(salary.max_salary_list)
print(salary.period_list)
print(salary.hour_salary_list)

min_salary_list = salary.min_salary_list
max_salary_list = salary.max_salary_list
period_list = salary.period_list
hour_salary_list = salary.hour_salary_list

df = salary.dataframe_salary()
print(df)
```

## Additional Data Cleaning

### The Days to Weeks systems conversion 
- Although it is not require to be saved in the relational database as it would appears as duplicated informations
- The conversion here is done for future data analysis references 
- A new column after convresions to weeks systems is stored as a column called Date of Job Listed By Weeks. 

```python
class ChangeToWeek:
    def __init__(self):
        pass

    def change_to_weeks(self, change_date):
        if 'day' in change_date and 'Apply easily' not in change_date:
            dayy = int(re.search(r'\d+', change_date).group())
            days_ago = dayy
            return round(days_ago / 7,2)
        elif 'month' in change_date:
            months_agoo = int(re.search(r'\d+', change_date).group())
            months_ago = months_agoo
            return round(months_ago * 4,2)  
        elif 'Apply easily' in change_date:
            return 1
        elif 'hour' in change_date:
            return 0  
        elif 'days ago' in change_date:
            days_agoo = int(re.search(r'\d+', change_date).group()) / 7
            return days_agoo
        else:
            return 0


weeks = ChangeToWeek()


df['Date of Job Listed By Weeks'] = df['Date Job Listed'].apply(weeks.change_to_weeks)

print("The Date Of Job Listed By Weeks ")
print(df['Date of Job Listed By Weeks'])
print(df)
```

# Matching Similar Jobs in United States. 
### The infomation is shown as below 

#### Large datasets require more times to do the matching 

1. The number of Groups of Specific Job 
2. The similar Job Title of the specific group
3. Job frequency for each group
4. The Companies of the specific Job group
5. The Salaries details of each similar Job in every specific groups 

```python

def string_similarity(job1, job2):
    job1_ = job1.lower()
    job2_ = job2.lower()
    return job1_ in job2_ or job1_ in job2_
  
jobs = []


for index , row in df.iterrows():
    matched_group_str = None

    for group in jobs:
        for job_title in group["job"]:
            if string_similarity(row["Job"], job_title):
                matched_group_str = group
                break
        if matched_group_str:
            break

    if matched_group_str is not None:
        matched_group_str["job"].append(row["Job"])
        matched_group_str["Count"] += 1
        matched_group_str["Companies"].add(row["Company"])
        matched_group_str["Salaries"].append({
            "salary": row["Salary"],
            "Average_Salary": row["Average Salary"],
            "Mininimum_Salary": row["Minimum Salary"],
            "Maximum_Salary": row["Maximum Salary"],
            "Hourly_Salary": row["Hourly Salary"],
            "Period": row["Period"]
        })
    else:
        jobs.append({
            "job": [row["Job"]],
            "Count": 1,
            "Companies": {row["Company"]},
            "Salaries": [{
                "salary": row["Salary"],
                "Average_Salary": row["Average Salary"],
                "Mininimum_Salary": row["Minimum Salary"],
                "Maximum_Salary": row["Maximum Salary"],
                "Hourly_Salary": row["Hourly Salary"],
                "Period": row["Period"]
            }]
        })


for x, group in enumerate(jobs):
    print(f"Group {x + 1}:")
    
    for index, job in enumerate(group["job"]):
        print(f" {x + 1} Job: {job}")
    
    print(f" ---Job Frequency---: {group['Count']}")
    
    print(" ---Companies---:")
    for company_info in group['Companies']:
        print(company_info)
    
    print(" ---Salaries---:")
    for salary_info in group['Salaries']:
        print(f"   -Salary: {salary_info.get('salary', 'Non Available')}")
        print(f"      Average_Salary: {salary_info.get('Average_Salary', 'Non Available')}")
        print(f"      Mininumum_Salary: {salary_info.get('Mininimum_Salary', 'Non Avaialble')}")
        print(f"      Maximum_Salary: {salary_info.get('Maximum_Salary', 'Non Available')}")
        print(f"      Hourly_Salary: {salary_info.get('Hourly_Salary', 'Non Avaialble')}")
        print(f"      Period: {salary_info.get('Period', 'Non Available')}")
```

# Database Schema

### The relational database should be stored in a normalized form so that it can be called the following queries as stated in the CA2 projects 

- List all jobs in a City/region
- List average (advertised) pay
- List all jobs requiring Python
- List jobs that are over 4 weeks old
- List jobs advertised by a specific company


### The Normalization Process 

The terms being used 
- UNF : Unormalized Form 
- 1NF : Frist Normalization Form 
- 2NF/3NF : Second Normalization Form / Third Normalization Form (The final normalized structure)



#### UNF 
These are all the available columns names 
- Job, Company, City, Date Job Listed, Minimum Salary Salary, Average Salary, Period, Date of Job Listed by Weeks

#### 2NF/3NF 
#### This is the final normalized table 
#### Table Company_Info and Job_Info

- Company_Info (Company_ID, Company Name, City)
- Job_Info (Company_ID, Job, Minimum Salary, Maximum Salary, Period, Date Job Listed)

```python
import mysql.connector
```

```python
import mysql.connector



try:
    conn = mysql.connector.connect(host="localhost", user="root", password="")
    mycursor = conn.cursor()
    
    # Create the database if it doesn't exist
    mycursor.execute("CREATE DATABASE IF NOT EXISTS CA2_OOP_TAN_PEI_WEN")
    mycursor.execute("USE CA2_OOP_TAN_PEI_WEN")
    
    # Create Company_Info table
    mycursor.execute("""
        CREATE TABLE IF NOT EXISTS Company_Info (
            Company_ID INT AUTO_INCREMENT PRIMARY KEY,
            Company VARCHAR(500),
            City VARCHAR(500)
            )
    """)
    
    # Create Job_Info table
    mycursor.execute("""
        CREATE TABLE IF NOT EXISTS Job_Info (
            Company_ID INT AUTO_INCREMENT,
            Job VARCHAR(500),
            Minimum_Salary INT,
            Maximum_Salary INT,
            Period VARCHAR(50),
            Date_Job_Listed VARCHAR(500),
            FOREIGN KEY (Company_ID) REFERENCES Company_Info(Company_ID)
            )
    """)
    
    # Insert data into Company_Info table
    companies_data = df[[ 'Company', 'City']].values.tolist()
    mycursor.executemany("INSERT INTO Company_Info (Company, City) VALUES (%s, %s)", companies_data)
    conn.commit()
    
    
    # Insert data into Job_Info  table
    job_data = df[['Job', 'Minimum Salary', 'Maximum Salary', 'Period', 'Date Job Listed']].values.tolist()
    mycursor.executemany("INSERT INTO Job_Info (Job, Minimum_Salary, Maximum_Salary, Period, Date_Job_Listed) VALUES ( %s, %s, %s, %s, %s)", job_data)
    conn.commit()
    
    
    #print data from Job_Info table
    print("---JOB INFORMATION----")
    mycursor.execute("SELECT * FROM Job_Info")
    for row in mycursor.fetchall():
        print(row)

    #print data from Company_Info table
    print("---COMPANY INFORMATION----")
    mycursor.execute("SELECT * FROM Company_Info")
    for row in mycursor.fetchall():
        print(row)


except mysql.connector.Error as e:
    print("Error {}: {}".format(e.errno, e.msg))
finally:
    if conn:
        conn.close()
    
```
